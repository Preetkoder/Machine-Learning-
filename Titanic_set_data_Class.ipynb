```python
#

def fill_age(row):
    pclass = row[0]
    gender = row[1]
    age = row[2]
    if pd.isnull(age):
        if pclass == 1:
            if gender == 'male':
                return 40
            else:
                return 35
        elif pclass==2:
            if gender == 'male':
                return 30
            else:
                return 28
        elif pclass == 3:
            if gender == 'male':
                return 25
            else:
                return 21
    else:
        return age
```

```python
def fill_null(Cabin):
    if pd.isnull(Cabin):
        return 0
    else:
        return 1
```

```python
# now we will do LAbel encoding as due to categorical data

from sklearn.preprocessing import LabelEncoder

label_ecoder = LabelEncoder()
```

```python
label_ecoder.fit(titanic['Gender'])
```

```python
titanic['Gender'] = label_ecoder.transform(titanic['Gender'])
```

```python
from sklearn.preprocessing import OneHotEncoder

one_hot_encoder = OneHotEncoder()

one_hot_encoder.fit(titanic[['Embarked']])
```

```python
one_hot_encoder.transform(titanic[['Embarked']])
```

```python
temp = one_hot_encoder.transform(titanic[['Embarked']]).toarray()
```

```python
one_hot_encoder.categories_
```

```python
temp_df = pd.DataFrame(temp,columns=['C', 'Q', 'S'],dtype = int)
```

```python
temp_df.head()
```

```python
titanic = pd.concat([titanic,temp_df],axis=1)
```

```python
titanic.drop('Embarked',axis=1,inplace=True)
```

```python
titanic.head()
```

```python
X = titanic.drop('Survived',axis=1)
Y = titanic['Survived']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression

logistic_regression = LogisticRegression()
```

```python
logistic_regression.fit(X_train,Y_train)
```

```python
predicitons = logistic_regression.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report

print(confusion_matrix(predicitons,Y_test))
```

```python
print(classification_report(predicitons,Y_test))
```
